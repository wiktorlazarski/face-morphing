# Helen dataset first look

```python
import os

import cv2
import matplotlib.pyplot as plt
```

```python
def show_sample(ann_file, img_file):
    KEYPOINTS = []

    with open(ann_file, "r") as ann_file:
        for i, line in enumerate(ann_file.readlines()):
            if i == 0:
                continue
        
            coords = [float(val) for val in line.split(",")]
            KEYPOINTS.append(tuple(coords))
    
    img = cv2.imread(img_file)

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    img_with_keypoints = img.copy()

    for keypoint in KEYPOINTS:
        x, y = round(keypoint[0]), round(keypoint[1])
        img_with_keypoints = cv2.circle(img_with_keypoints, (x, y), radius=5, color=(0,255,0), thickness=-1)
    
    f, axis = plt.subplots(1, 2, figsize=(24, 10))
    
    axis[0].imshow(img)
    axis[1].imshow(img_with_keypoints)
```

```python
show_sample(ann_file="1.txt", img_file="100032540_1.jpg")
```

```python
show_sample(ann_file="2.txt", img_file="100040721_1.jpg")
```
